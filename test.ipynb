```python
import matplotlib.pyplot as plt

#Hypothetical network traffic data
days = ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"]
traffic = [2, 5, 1, 8, 4, 7, 3,]
threshold = 6  #Danger threshold

plt.figure(figsize=(10, 5))

plt.plot(days, traffic, marker='o', linestyle='-', color='b', label='Daily Traffic')

plt.axhline(y=threshold, color='r', linestyle='--', label='Danger Threshold')

for i in range(len(traffic)):
    if traffic[i] > threshold:
        plt.plot(days[i], traffic[i], 'ro')
        
        plt.title("Network Traffic Analysis With Anomaly Threshold")
        plt.xlabel("Days of the week")
        plt.ylabel("Number of requests / attacks")
        plt.legend()
        plt.grid(True, linestyle=':', alpha=0.6)
        
        plt.show();
```
